```python
import pandas as pd
import plotly.express as px

# Read the CSV file into a Pandas DataFrame
Job_data = pd.read_csv(r"C:\Users\ASUS\Desktop\FILES\Data analysis\projects\Job db.csv")

Job= (
    Job_data
    .groupby(["Experience_categories","Gender"], as_index=False)["Salary"].mean()
    .sort_values("Salary", ascending=True)
)
Job.head()
```

```python
#Line Chart
fig = px.line(
    Job.query("Gender in ['Male', 'Female']"),
    x="Experience_categories",
    y="Salary",
    color="Gender",
    title="Years of Experience v/s Salary (on the basis of Gender)"
)

# Display the plot
fig.show()

```

```python

import pandas as pd
import plotly.express as px

Job_data = pd.read_csv(r"C:\Users\ASUS\Desktop\FILES\Data analysis\projects\Job db.csv")

Job= (
    Job_data
    .groupby(["Job_Title","Years_of_Experience"], as_index=False)["Salary"]
    .mean()
    .sort_values("Years_of_Experience", ascending=False)
)

Job.head()
```

```python
#Scatter Plot

fig = px.scatter(
    Job, 
    "Years_of_Experience", 
    "Salary", 
    color="Job_Title", 
    title="Scatter Plot",
)
fig.show()
```

```python
#Bubble chart
fig = px.scatter(
    Job, 
    "Years_of_Experience", 
    "Salary", 
    color="Job_Title", 
    title="Bubble chart",
    size="Years_of_Experience"
)
fig.show()
```
